```python
import os
import numpy as np
import glob
import pdb
import matplotlib.pyplot as plt
```

```python
train_set=[]
test_set=[]
with open("./logfile.txt", "r+") as f:
    for line in f:
        if "=>" in line:
            train=float(line.split(':')[-1].rstrip('\n').rstrip(' '))
            train_set.append(train)
        if "Test" in line:
            test=float(line.split(':')[-1].rstrip('\n').rstrip(' '))
            test_set.append(test)
            
nptrain=np.array(train_set)
nptest=np.array(test_set)

plt.figure()
plt.plot(np.arange(18),nptrain[:18])
plt.plot(np.arange(18),nptest[:18])
plt.xlabel('epoch')
plt.ylabel('error')
plt.legend(['train','valid'])
plt.show()
```
