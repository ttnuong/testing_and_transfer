```python
import os
import numpy as np
import glob
import pdb
import matplotlib.pyplot as plt
```

```python
train_set=[]
test_set=[]
train_set_SSIM=[]
test_set_SSIM=[]
train_set_MSE=[]
test_set_MSE=[]
with open("C:\\Users\\Nuong\\Desktop\\heidelberg\\MA-Arbeitsordner\\experiment_dir\\20190510-084116_test_nuong\\log\\logfile.txt", "r+") as f:
    for line in f:
        if "Mean Loss" in line:
            train=float(line.split(':')[-1].rstrip('\n').rstrip(' '))
            train_set.append(train)
        if "Average loss:" in line:
            test=float(line.split(':')[-1].rstrip('\n').rstrip(' '))
            test_set.append(test)
        if "Mean MSE Loss" in line:
            train=float(line.split(':')[-1].rstrip('\n').rstrip(' '))
            train_set_MSE.append(train)
        if "Average MSE loss:" in line:
            test=float(line.split(':')[-1].rstrip('\n').rstrip(' '))
            test_set_MSE.append(test)
        if "Mean SSIM Loss" in line:
            train=float(line.split(':')[-1].rstrip('\n').rstrip(' '))
            train_set_SSIM.append(train)
        if "Average SSIM loss:" in line:
            test=float(line.split(':')[-1].rstrip('\n').rstrip(' '))
            test_set_SSIM.append(test)
            
nptrain=np.array(train_set)
nptest=np.array(test_set)

plt.figure(figsize=(8,8))
plt.subplot(121)
plt.plot(np.arange(len(nptrain)),nptrain)
plt.plot(np.arange(len(nptrain)),nptest)
plt.xlabel('epoch')
plt.ylabel('error')
plt.legend(['train','valid'])
plt.subplot(122)
plt.plot(np.arange(len(nptrain)),np.log(nptrain))
plt.plot(np.arange(len(nptrain)),np.log(nptest))
plt.xlabel('epoch')
plt.ylabel('error')
plt.legend(['train','valid'])
plt.show()
```

## validation is lower than training
some people say that the validation test set error could be lower in some cases

The dropout layer introduces noise that is not injected during the test period. 

 error in your test set will be lower due to better generalization

```python
plt.figure(figsize=(8,8))
plt.subplot(121)
plt.plot(np.arange(len(train_set_MSE)),np.array(train_set_MSE))
plt.plot(np.arange(len(nptest)),np.array(test_set_MSE))
plt.xlabel('epoch')
plt.ylabel('error')
plt.legend(['train','valid'])
plt.title('MSE loss')
plt.subplot(122)
plt.plot(np.arange(len(train_set_MSE)),np.log10(np.array(train_set_MSE)))
plt.plot(np.arange(len(nptest)),np.log10(np.array(test_set_MSE)))
plt.xlabel('epoch')
plt.ylabel('error')
plt.legend(['train','valid'])
plt.title('MSE logLoss')
plt.show()

plt.figure(figsize=(8,8))
plt.subplot(121)
plt.plot(np.arange(len(train_set_MSE)),np.array(train_set_SSIM))
plt.plot(np.arange(len(nptest)),np.array(test_set_SSIM))
plt.xlabel('epoch')
plt.ylabel('error')
plt.legend(['train','valid'])
plt.title('SSIM loss')
plt.subplot(122)
plt.plot(np.arange(len(train_set_MSE)),np.log10(np.array(train_set_SSIM)))
plt.plot(np.arange(len(nptest)),np.log10(np.array(test_set_SSIM)))
plt.xlabel('epoch')
plt.ylabel('error')
plt.legend(['train','valid'])
plt.title('SSIM logLoss')
plt.show()
```
